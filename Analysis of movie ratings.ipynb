```python
by_genre = movie_rating[['rating', 'genres', 'num ratings']]
by_genre.head()
```

```python
genre_count_copy = by_genre.copy()
for i in range(len(genre_count_copy['genres'])):
    if '|' in genre_count_copy['genres'][i]:
        genre_count_copy.drop(index = i, inplace = True)

genre_count_copy.size
```

```python
text = 'Adventure|Children|Fantasy'
list = text.split('|')
list
```

```python
by_genre['genres']
```

```python
for i in range(len(by_genre)):
    if '|' in by_genre.loc[i]['genres']:
        list_genre = by_genre.loc[i]['genres'].split('|')
        list_rating = []
        list_num_ratings = []
        for i in range(len(list_genre)):
            list_rating.append(by_genre.loc[i]['rating'])
            list_num_ratings.append(by_genre.loc[i]['num ratings'])
        dict = {'rating' : list_rating, 'genres' : list_genre, 'num ratings': list_num_ratings}
        df0 = pd.DataFrame(dict)
        genre_rating = pd.concat([genre_count_copy, df0], ignore_index = True)   
```

```python
genre_rating['rating*num_rating'] = genre_rating['rating']*genre_rating['num ratings']
genre_rating.head()
```

```python
num_movies = pd.DataFrame(genre_rating['genres'].value_counts()).reset_index()
num_movies.head()
```

```python
genre_rating[['genres', 'num ratings', 'rating*num_rating']].groupby('genres').sum().reset_index()
```

```python
rating_by_genre = pd.DataFrame()
```

```python
rating_by_genre = genre_rating[['genres', 'num ratings', 'rating*num_rating']].groupby('genres').sum().reset_index()
rating_by_genre['ave rating'] = rating_by_genre['rating*num_rating']/rating_by_genre['num ratings']
rating_by_genre = rating_by_genre.merge(num_movies, on = 'genres')
rating_by_genre.head()
```

```python
rating_by_genre = rating_by_genre.rename(columns = {'count': 'num movies'})
rating_by_genre = rating_by_genre.drop(columns = 'rating*num_rating')
rating_by_genre
```

```python
rating_by_genre.to_csv('Genre_Rating.csv')
```

```python
fig = plt.figure(figsize = (10, 20))
rating_by_genre[['genres', 'num ratings']].plot.bar(x = 'genres', y = 'num ratings', color = 'orange')
plt.xlabel('Genres of movies')
plt.ylabel('Number of Ratings')
plt.title('Numbers of ratings of genres of movies')
```

```python
fig = plt.figure(figsize = (10, 20))
rating_by_genre[['genres', 'ave rating']].plot.bar(x = 'genres', y = 'ave rating', color = 'red')
plt.xlabel('Genres of movies')
plt.ylabel('Average rating')
plt.title('Average ratings of genres of movies')
```
